```python
#equilibrium index 
arr=[12,14,15,16,15,13,13]
def equilibrium_index(arr):
    right_sum=sum(arr)
    left_sum=0
    for i in range(len(arr)):
        right_sum-=arr[i]
        if left_sum==right_sum:
           return i
        left_sum+=arr[i]
    return -1

result=equilibrium_index(arr)
print(result)
```

```python
#practice
arr=[1,3,5,7,8,7,9]
def equilibrium_index(arr):
    right_sum=sum(arr)
    left_sum=0
    for i in range(len(arr)):
        right_sum-=arr[i]
        if left_sum==right_sum:
            return i
        left_sum+=arr[i]
    return -1

result=equilibrium_index(arr)
print("equilibrum index:",result)
print("element on equilibrium index:",arr[result])
```
