```python
#ques-3 Situation of Person
temperature=float(input("enter your temperature:"))
if temperature>=85.0 and temperature<=91.0:
    print("Serious Hypothermia")
if temperature>91.0 and temperature<95.0:
    print("Mild Hypothermia")
if temperature>=95.0 and temperature<=98.0:
    print("Normal Temperature")
if temperature>98.0 and temperature<=100.0:
    print("Mild Fever")
if temperature>100.0 and temperature<=105.0:
    print("High Fever")
```

```python
#ques-4 Situation Description
N=int(input("enter a number"))
if N>0 and N%2==0:
    print("Number is positive even")
if N>0 and N%2!=0:
    print("Number is positive odd")
if N<0 and N%2==0:
    print("Number is negative even")
if N<0 and N%2!=0:
    print("Number is negative odd")
```

```python
#ques-9 To check if it is prime or not
x=int(input("enter a number"))
count=0
for i in range(1,x+1):
    if x%i==0:
        count+=1
if count==2:
    print("it is a prime number")
else:
    print("Not a prime number")
        

```
