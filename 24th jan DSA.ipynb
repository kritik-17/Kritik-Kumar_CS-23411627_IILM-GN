```python
#merging pattern 
n=int(input("enter the number of rows:"))
i=1
while i<=n:
    j=1
    while j<=i:
        print("*",end=' ')
        j+=1
    k=1
    while k<=i:
        print(k,end=' ')
        k+=1
    print()
    i+=1
        
```

```python
#merging 3 patterns
n=int(input("enter the number of rows to be printed:"))
i=1
while i<=n:
    #stars
    j=1
    while j<=i:
        print("*",end=' ')
        j+=1
    #numbers
    k=1
    while k<=i:
        print(k,end="")
        k+=1
    #alphabets
    l=1
    while l<=i:
        print("A",end="")
        l+=1
    print()
    i+=1
        
      
```

```python
#advance patterns
#pyramids
n=int(input("enter the number of rows to be printed:"))
i=1
while i<=n:
    #spaces
    s=1
    while s<=n-i:
        print(" ",end='')
        s+=1
        #stars
    j=1
    while j<=i:
        print("*",end=' ')
        j+=1
    print()
    i+=1
        
```

```python
#pattern qustion
n=4
i=1
while i<=n:
    j=1
    while j<=n-i:
        print(" ",end=' ')
        j+=1
    k=1
    while k<(2*i):
        print(k,end=' ')
        k+=1
    print()
    i+=1
```
