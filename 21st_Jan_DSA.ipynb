```python
#ques-1 HCF of two numbers
x=int(input())
y=int(input())
maxvalue=min(x,y)
ans=0
for i in range(1,maxvalue+1):
    if(x%i==0)and(y%i==0):
        ans=i
print(ans)
```

```python
#ques-2 LCM of two numbers
x=int(input())
y=int(input())
minvalue=max(x,y)
maxvalue=x*y
for i in range(minvalue,maxvalue+1):
    if i%x==0 and i%y==0:
        print(i)
        break #to stop the loop otherwise it will double every value
```
