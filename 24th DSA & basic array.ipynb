```python
n=int(input())
i=0
while i<n:
    j=0
    while j<n:
        print("*",end=" ")
        j+=1 #j=j+1
    i+=1
    print()
```

```python
n=int(input())
i=1
while i<=n:
    j=1
    while j<=i:
        print("*",end=" ")
        j+=1
    i+=1
    print()
```

```python
#swapping in python
lst = [12,13,14,15]
l=len(lst)
for i in range(l//2):
    lst[i],lst[l-i-1]=lst[l-i-1],lst[i]
print(lst)
```
